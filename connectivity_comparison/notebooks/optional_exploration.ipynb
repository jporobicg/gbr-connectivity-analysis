# Optional Exploration

Use this notebook after running `run_analysis.py` to inspect the saved outputs.

```python
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt

PROJECT_DIR = Path.cwd().parent
TABLES_DIR = PROJECT_DIR / 'outputs' / 'tables'
FIGURES_DIR = PROJECT_DIR / 'outputs' / 'figures'
```

```python
comparison = pd.read_csv(TABLES_DIR / 'per_time_matrix_comparison.csv')
comparison_all = comparison[comparison['subset'] == 'all'].sort_values('time_index')
comparison_all[['time_label', 'pearson_r', 'rmse', 'mae', 'frobenius_norm']].head()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(comparison_all['time_label'], comparison_all['pearson_r'], marker='o')
ax.set_ylabel('Pearson r')
ax.set_title('Single vs Bootstrap Mean Through Time')
ax.tick_params(axis='x', rotation=45)
ax.grid(alpha=0.3)
plt.tight_layout()
```
